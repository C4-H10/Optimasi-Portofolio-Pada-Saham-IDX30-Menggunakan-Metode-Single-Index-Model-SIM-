```python
# Dataframe Return IHSG (Log Return)
return_IHSG = np.log(ihsg / ihsg.shift(1))
return_IHSG = return_IHSG.dropna()
return_IHSG.head()
```

#PERHITUNGAN BOBOT PORTOFOLIO

### **Penjelasan Variabel**

1.  **RATA RATA / E(Ri)**: Tingkat pengembalian yang diharapkan (*expected return*) dari masing-masing saham, dihitung dari rata-rata imbal hasil historis selama periode pengamatan.
2.  **VARIAN**: Risiko total saham yang mencerminkan besarnya fluktuasi imbal hasil harian terhadap nilai rata-ratanya.
3.  **BETA ($\beta$)**: Indikator risiko sistematis yang mengukur sensitivitas imbal hasil saham terhadap pergerakan pasar (IHSG).
4.  **ALFA ($\alpha$)**: Komponen imbal hasil yang bersifat independen atau tidak dipengaruhi oleh pergerakan pasar secara keseluruhan.
5.  **VAR RESIDUAL ($\sigma_{ei}^2$)**: Risiko unik perusahaan (risiko tidak sistematis) yang dapat diminimalisir melalui diversifikasi.
6.  **ERB (Excess Return to Beta)**: Rasio kelebihan imbal hasil di atas suku bunga bebas risiko per unit risiko sistematis. Digunakan sebagai dasar utama pemeringkatan saham.
7.  **Ai & Bi**: Variabel antara yang mengintegrasikan nilai *excess return*, beta, dan risiko residual untuk menentukan titik pembatas.
8.  **Ci (Cut-off Point)**: Nilai batas akumulatif. Saham yang terpilih masuk portofolio optimal adalah saham yang memiliki nilai **ERB > Ci**.
9.  **Zi**: Koefisien bantuan untuk menentukan proporsi dana, yang mempertimbangkan selisih ERB terhadap nilai *cut-off* ($C^*$) dan risiko residual.
10. **Bobot (Wi)**: Proporsi alokasi modal (persentase dana) yang diinvestasikan pada masing-masing saham dalam portofolio sehingga total bobot adalah 100%.

```python
market_return = return_IHSG.iloc[:, 0]
var_market_sample = market_return.var(ddof=1)
var_market_pop = market_return.var(ddof=0)

rata_rata_suku_bunga = np.mean(suku_bunga_list)
rf = rata_rata_suku_bunga / len(return_saham)
print(f"Rata-rata harian suku bunga BI : {rf}")

betas = {}
for col in return_saham.columns:
    cov_p = np.cov(return_saham[col], market_return, ddof=0)[0][1]
    betas[col] = cov_p / var_market_sample

beta_series = pd.Series(betas)
rata_rata = return_saham.mean()
varian_saham = return_saham.var(ddof=0)
varian_residual = varian_saham - (beta_series**2 * var_market_pop)

ai = (rata_rata - (((rata_rata - rf) / beta_series))) / varian_residual
bi = (beta_series**2) / varian_residual

summary_df = pd.DataFrame({
    'RATA RATA': rata_rata,
    'VARIAN': varian_saham,
    'SUM': return_saham.sum(),
    'BETA': beta_series,
    'ALFA': rata_rata - (beta_series * market_return.mean()),
    'E(Ri)': rata_rata,
    'ERB': (rata_rata - rf) / beta_series,
    'VAR RESIDUAL': varian_residual,
    'Ai': ai,
    'Bi': bi
})

summary_df.index.name = 'Saham'
summary_df = summary_df.sort_values(by='ERB', ascending=False) #sort berdasarkan ERB
summary_df['Ai Kumulatif'] = summary_df['Ai'].cumsum()
summary_df['Bi Kumulatif'] = summary_df['Bi'].cumsum()
summary_df['Ci'] = (var_market_pop * summary_df['Ai Kumulatif']) / (1 + (var_market_pop * summary_df['Bi Kumulatif']))
summary_df['ERB - Ci'] = summary_df['ERB'] - summary_df['Ci']
summary_df['Zi'] = (summary_df['BETA']/summary_df['VAR RESIDUAL'])*summary_df['ERB - Ci']
summary_df = summary_df[summary_df['ERB - Ci'] > 0]
summary_df['Bobot (Wi)'] = summary_df['Zi']/summary_df['Zi'].sum()
summary_df
```

```python
print(f"Jumlah saham yang terpilih sebanyak : {len(summary_df)} saham")
print(summary_df[["Bobot (Wi)"]].sort_values(by="Bobot (Wi)", ascending=False))
```

#KINERJA PORTOFOLIO (INDEKS TREYNOR)

**Ekspektasi Return Portofolio** : <br>
rata-rata tingkat keuntungan yang diharapkan setiap hari
<br><br>
**Beta Portofolio** : <br>
Jika beta > 1 portofolio lebih volatil dibandingkan pasar.<br>
Jika beta < 1 portofolio lebih stabil. <br>
Jika beta = 1 maka portofolio bergerak mengikuti pergerakan pasar (IHSG).
<br><br>
**Indeks Treynor** :<br>
Jika Indeks Treynor portofolio > Indeks Treynor IHSG, maka portofolio yang disusun sudah optimal


```python
# Rata-rata Return IHSG
rata_rata_ihsg = return_IHSG.mean().iloc[0]

# Suku Bunga Bebas Risiko (Rf)
risk_free = rf

# Beta IHSG
beta_market = 1.0

# Rumus Treynor IHSG
treynor_ihsg = (rata_rata_ihsg - risk_free) / beta_market

# Ekspektasi Return Portofolio
return_portofolio = (summary_df['Bobot (Wi)'] * summary_df['E(Ri)']).sum()

# Beta Portofolio
beta_portofolio = (summary_df['Bobot (Wi)'] * summary_df['BETA']).sum()

# Rumus Treynor Portofolio
treynor_portofolio = (return_portofolio - risk_free) / beta_portofolio

print("=== KINERJA PORTOFOLIO (INDEKS TREYNOR) ===")
print(f"Return Harian IHSG         : {rata_rata_ihsg:.6f}")
print(f"Return Harian Portofolio   : {return_portofolio:.6f}")
print("-" * 45)
print(f"Beta Portofolio            : {beta_portofolio:.6f}")
print("-" * 45)
print(f"Indeks Treynor IHSG        : {treynor_ihsg:.6f}")
print(f"Indeks Treynor Portofolio  : {treynor_portofolio:.6f}")
print("-" * 45)

if treynor_portofolio > treynor_ihsg:
    print("KESIMPULAN: Portofolio OPTIMAL")
else:
    print("KESIMPULAN: Portofolio BELUM OPTIMAL")
```
